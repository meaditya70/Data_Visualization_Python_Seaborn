# Data Visualization using Seaborn library - Python

>Here we will be discussing *<b>What</b> are the differents types of plots, *<b>how</b> to use them and *<b>when</b> to use them

Types of Plots:
    1. Line Plot
    2. Bar Plot
    3. Scatter Plot
    4. LM Plot -> regression line + scatter plot
    5. Swarm Plot, Box Plot and Violin Plot.
    6. Histogram
    7. Density plots - 1D, 2D.
    8. HeatMaps - for Missing values.
    
Using MatPlotLib Library:
    9. Merging Multiple PLots
    10. Sizing/resizing the plots, labels and legends.

```python
import pandas as pd
import seaborn as sns
```

```python
dataset=pd.read_csv("IRIS.csv")
dataset.head()
```

```python
dataset.drop(['species'],axis=1,inplace=True)
dataset.head()
```

# Line plot

To view variation with respect to time.<br>
<i> It shows trends with time.</i>

```python
sns.lineplot(data=dataset)
```

# Bar Plot
To see the average variation column wise.<br>
<i> It shows relationship with columns.</i>

```python
sns.barplot(data=dataset)
```

```python
import numpy as np
np.mean(dataset.sepal_length)
```

# Scatter Plot
To view the spread of data.<br>
Shows relation between two continuous variables, can be color-coded with a third categorical variable.

```python
sns.scatterplot(x=dataset.sepal_length,y=dataset.petal_length,hue=dataset.species)
```

# LM Plot
To get regression line along with scatter plots.<br>
There can be multiple regression line wrt.hue. <br>

```python
sns.lmplot(x="sepal_length",y="petal_length",hue="species",data=dataset)
```

```python
sns.lmplot(x="sepal_length",y="petal_length",data=dataset)
```


> <b>Reg Plot</b> <br>
Just a subset of LM plot, we cannot separate data based on hue.<br>
Hence, we can only get a single regression line.

```python
sns.regplot(x="sepal_length",y="petal_length",data=dataset)
```

# Swarm Plot, Box Plot, Violin Plot
>Used to plot categorical data. <br>
One axis is categorical and other must be continuous. 

```python
sns.swarmplot(x=dataset.species,y=dataset.sepal_length)
```

```python
sns.boxplot(x=dataset.species,y=dataset.sepal_length)
```

```python
sns.swarmplot(x=dataset.species,y=dataset.sepal_length,color='black')
sns.boxplot(x=dataset.species,y=dataset.sepal_length)
```

```python
sns.violinplot(x=dataset.species,y=dataset.sepal_length)
```

```python
sns.swarmplot(x=dataset.species,y=dataset.sepal_length,color='black')
sns.violinplot(x=dataset.species,y=dataset.sepal_length)
```

```python
sns.swarmplot(x=dataset.species,y=dataset.sepal_length,color='black')
sns.boxplot(x=dataset.species,y=dataset.sepal_length)
sns.violinplot(x=dataset.species,y=dataset.sepal_length)
```

# Histograms and Density Plots
> Histograms - To see the distibution of a column values in a Dataset. <br>
Density Plots - Smoothened Histograms.

```python
mybins=[3,4,5,6,7,8,9]
sns.distplot(a=dataset.sepal_length,kde=False,bins=mybins)
```

```python
sns.kdeplot(data=dataset.sepal_length)
```

### 2D KDE Plot

```python
sns.jointplot(x=dataset.sepal_length,y=dataset.petal_length)
```

```python
sns.jointplot(x=dataset.sepal_length,y=dataset.petal_length,kind='kde')
```
